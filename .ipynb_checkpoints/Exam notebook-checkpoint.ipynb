# Preprocessing

```python
### Preprocessing
# remove na's in content column 
dtf.dropna(subset=['content'], inplace=True)

# prepare dataframe
content = dtf["content"].tolist()
```

```python
# stopwords removal 
from nltk.corpus import stopwords
stopwords=set(stopwords.words('english'))

def remove_stopwords(data):
    output_array=[]
    for sentence in data:
        temp_list=[]
        for word in sentence.split():
            if word.lower() not in stopwords:
                temp_list.append(word)
        output_array.append(' '.join(temp_list))
    return output_array

xcontent = remove_stopwords(content)
```

```python
import spacy
from spacy import displacy
import en_core_web_sm
nlp = en_core_web_sm.load()

result = []

s = xcontent
for i in s:
    doc = nlp(i)
    newString = i
    for e in reversed(doc.ents): 
        start = e.start_char
        end = start + len(e.text)
        newString = newString[:start] + e.label_ + newString[end:]
    result.append(newString)
```

```python
dtf['ner_content'] = result
```

```python
y = dtf["Y"]
Ner_X = CountVectorizer().fit_transform(result)
```

```python
# split data
X_train, X_test, y_train, y_test = model_selection.train_test_split(Ner_X,y, train_size=0.70,test_size=0.30,random_state = 44)
```

```python
# create Gaussian Naive Bayes model
model = MultinomialNB()
#X_train = CountVectorizer().fit_transform(X_train)
#X_train = X_train.toarray()
#y_train = CountVectorizer().fit_transform(y_train)
#y_train = y_train.toarray()
classifier = model.fit(X_train, y_train)
```

```python
print('Training set score: {:.4f}'.format(model.score(X_train, y_train)))

print('Test set score: {:.4f}'.format(model.score(X_test, y_test)))
```

```python
# predictions
expected = y_test
predicted = model.predict(X_test)

# accuracy and stats
print(metrics.classification_report(expected, predicted))
print(metrics.confusion_matrix(expected, predicted))
```

```python
print("Accuracy:",metrics.accuracy_score(y_test, predicted))
```

```python
from sklearn.metrics import plot_confusion_matrix
disp = plot_confusion_matrix(classifier, X_test, y_test,
                                 cmap=plt.cm.Blues)

print(disp.confusion_matrix)

plt.show()
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted)

TP = cm[0,0]
TN = cm[1,1]
FP = cm[0,1]
FN = cm[1,0]

```

```python
precision = TP / float(TP + FP)


print('Precision : {0:0.4f}'.format(precision))
```

```python
recall = TP / float(TP + FN)

print('Recall or Sensitivity : {0:0.4f}'.format(recall))
```
